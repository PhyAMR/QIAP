```python
import numpy as np
from numpy import linalg as LA
from scipy import linalg as SLA

import matplotlib
from matplotlib import pyplot as pl
```

```python
Eta=0.8
print('Value of Eta =>', Eta)
```

```python
Hamiltonian=np.array([[0,Eta,0],[Eta,1,Eta],[0,Eta,0]])
print(Hamiltonian)
```

```python
def TimeEvo(time):
    return(SLA.expm(-1j*Hamiltonian*time))

TimeEvo(0)
```

```python
def Probs(time):
    return(np.absolute(np.dot(TimeEvo(time),np.array([1,0,0])))**2)

Probs(42)
```

```python
time_list = np.linspace(0,4.2/Eta**2,1000)
```

```python
p_list = np.array([Probs(time) for time in time_list])
```

```python
matplotlib.rcParams['figure.figsize'] = [12,9]
pl.figure()
pl.plot(time_list, p_list)
#pl.plot(time_list, p_list[:,0])
#pl.plot(time_list, p_list[:,1], '--')
#pl.plot(time_list, p_list[:,2], '-.')
pl.show()
```
